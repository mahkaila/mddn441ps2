### Edge Detection

```python
command = "convert inputs/dog.jpg \
  -crop 640x480+1500+1680 \
  -resize 320x240 \
  -edge 2 \
  temp/edge_test.png"
print(command)
!{command}
Image(filename='temp/edge_test.png') 
```

```python
command = "convert britney/thumb0831.jpg \
  -colorspace gray \
  -resize 320x240 \
  -sketch 0x15+150 \
  temp/sketch_test2.png"
print(command)
!{command}
Image(filename='temp/sketch_test2.png') 
```

```python
command = "convert britney/thumb0831.jpg \
  -colorspace gray \
  -resize 320x240 \
  -fuzz 10% \
  -transparent white \
  -alpha extract \
  temp/sketch_test3.png"
print(command)
!{command}
Image(filename='temp/sketch_test3.png')
```

```python
command = "convert britney/thumb0831.jpg temp/sketch_test2.png temp/sketch_test3.png \
-resize 320x240 \
-composite \
temp/sketch_test4.png"
print(command)
!{command}
Image(filename='temp/sketch_test4.png')
```

```python
for i in range(50, 100):
    infile = "britney/thumb{:04d}.jpg".format(i)
    outfile = "britvid/brit_{:04d}.png".format(i)
    alphafile = "britalpha/alpha_{:04d}.png".format(i)
    command = "convert {} \
      -colorspace gray \
      -resize 320x240 \
      -sketch 0x15+150 \
      {}".format(infile, outfile)
    print(command)
    !{command}
```

```python
for i in range(50, 500):
    infile = "britney/thumb{:04d}.jpg".format(i)
    outfile = "britvid/brit_{:04d}.png".format(i)
    alphafile = "britalpha/alpha_{:04d}.png".format(i)
    compfile = "britcomp/comp_{:04d}.png".format(i)
    command = "convert {} \
      -colorspace gray \
      -resize 320x240 \
      -sketch 0x15+150 \
      {}".format(infile, outfile)
    print(command)
    !{command}
    command2 = "convert {} \
      -colorspace gray \
      -resize 320x240 \
      -fuzz 10% \
      -transparent white \
      -alpha extract \
      {}".format(infile, alphafile)
    print(command2)
    !{command2}
    command3 = "convert {} {} {} \
        -resize 320x240 \
        -composite \
        {}".format(infile, outfile, alphafile, compfile)
    print(command3)
    !{command3}
    
```
